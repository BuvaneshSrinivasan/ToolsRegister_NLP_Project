### Plot N-grams

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
```

```python
def plot_Ngram(ng_tokens, header, Ngram_type=1, top=25):
  dic_words_freq = []
  if  (Ngram_type==1): dic_words_freq = nltk.FreqDist(ng_tokens)
  else: dic_words_freq = nltk.FreqDist(nltk.ngrams(ng_tokens, Ngram_type))
  df_uni = pd.DataFrame(dic_words_freq.most_common(), columns=["Word","Freq"])
  df_uni.set_index("Word").iloc[:top,:].sort_values(by="Freq").plot(kind="barh", title=header, legend=False).grid(axis='x')
```

```python
accComment_corpus = df['Description']
accLevel_tokens = nltk.tokenize.word_tokenize(accComment_corpus.str.cat(sep=" "))
plot_Ngram (accLevel_tokens, "Unigrams", top=25)
plot_Ngram (accLevel_tokens, "Bigrams", Ngram_type=2, top=25)
plot_Ngram (accLevel_tokens, "Trigrams", Ngram_type=3, top=25)
```

### Preprocessing description

```python
def text_processing(text):
  text = re.sub('\[[^]]*\]', '', text)  #Removing the square brackets
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  #Removing special character and keep only words and numbers
  text_tokens = word_tokenize(text)
  filtered_text = [w.lower() for w in text_tokens if not w.lower() in stop_words]  #Removing stop words
  lemmatizer = WordNetLemmatizer()
  lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_text]
  final_text = " ".join(lemmatized_words)
  return final_text
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def cos_sim(sentence1_emb, sentence2_emb):
    cos_sim = cosine_similarity(sentence1_emb, sentence2_emb)
    return np.diag(cos_sim)
```

```python
def compute_cosine_similarity(sentence1_emb, sentence2_emb):
    cos_sim_values = cosine_similarity(sentence1_emb.reshape(1, -1), sentence2_emb.reshape(1, -1))
    return cos_sim_values[0, 0]
```

```python
def compute_sentence_embedding(sentence, model):
    words = sentence.split()
    valid_words = [word for word in words if word in model]
    if len(valid_words) > 0:
        return np.mean([model[word] for word in valid_words], axis=0)
    else:
        # If no valid words are found, return a zero vector
        return np.zeros(model.vector_size)
```
